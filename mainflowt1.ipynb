```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step-1 : Data Loading
df = pd.read_csv("C:/Users/HP/Desktop/student-mat.csv", sep=';', quotechar='"')
print("Data Loaded successfully")
print(df.head()) # Display the first few rows

# Step-2 : Data Exploration
# Check for missing values
print("\nCheck for Missing Values:")
print(df.isnull().sum()) 

# Display the datatypes
print("\nThe datatypes :")
print(df.dtypes)

# Size of the dataset
print("\nSize of the dataset :")
print(df.shape)

# Step-3 : Data Cleaning
# Remove duplicates
print("\nDuplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())

# Step-4 : Data Analysis
# Question-1 : What is the average score in math (G3)?
average_score_G3 = np.mean(df['G3'])
print("\nAverege Math Score (G3) : ", average_score_G3)

# Question-2 : How many students scored above 15 in their final grade (G3)?
above_15_count = df[df['G3']>15].shape[0]
print("\nNumber of students scoring above 15 : ",above_15_count)

# Question-3 : Is there a correlation between study time (study time) and the final grade (G3)?
correlation = df['studytime'].corr(df['G3'])
print("\nCorrelation between study time and final grade",correlation)

# Question-4 : Which gender has a higher average final grade (G3)?
gender_avg = df.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender : ",gender_avg)

# Step-5 : Data Visualization
# Histogram of final grades
plt.figure()
plt.hist(df['G3'],color='black', edgecolor='white')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()

# Scatter plot of study time vs final grade
plt.figure()
sns.scatterplot(x='studytime', y='G3', data=df,hue='sex')
plt.title("Study Time vs Final Grade")
plt.xlabel("Weekly Study Time")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()

# Bar chart of average scores by gender
plt.figure()
sns.barplot(x='sex', y='G3', data=df, )
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```
